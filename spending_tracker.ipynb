```python
data['Impact√© √† Caps'] = data['Impact√© √† Caps'].abs()
data['Impact√© √† Emilian'] = data['Impact√© √† Emilian'].abs()
data.head(2)
```

```python
# Fonction pour extraire l'ann√©e manuellement
def extract_year_manual(date_str):
    # S√©parer la cha√Æne en composants jour, mois, ann√©e et heure:minute
    components = date_str.split()
  
    year_str = components[2]
    year = int(year_str)
    return year
```

```python
# Fonction pour extraire le mois manuellement avec un mapping fran√ßais -> anglais
def extract_month_manual(date_str):
    # S√©parer la cha√Æne en composants jour, mois, ann√©e et heure:minute
    components = date_str.split()
    
    month_str = components[1]
      
    month = month_str.lower()
    
    return month
```

```python
# Appliquer les fonctions aux colonnes 'Date & heure'
data['Ann√©e'] = data['Date & heure'].apply(extract_year_manual)
data['Mois'] = data['Date & heure'].apply(extract_month_manual)

# Dictionnaire pour mapper les noms de mois en num√©ros de mois
mois_map = {
    'janvier': 1, 'f√©vrier': 2, 'mars': 3,
    'avril': 4, 'mai': 5, 'juin': 6,
    'juillet': 7, 'ao√ªt': 8, 'septembre': 9,
    'octobre': 10, 'novembre': 11, 'd√©cembre': 12
}

# Ajouter une colonne 'Num√©ro de Mois'
#df['Num√©ro de Mois'] = df['Mois'].map(mois_map)

# Convertir les colonnes Ann√©e et Num√©ro de Mois en une colonne Date
data['Date'] = pd.to_datetime(data['Ann√©e'].astype(str) + '-' + (data['Mois'].map(mois_map)).astype(str) + '-01')
data.head(5)
```

```python
columns = ["Date", "Ann√©e", "Mois", "Type de transaction", "Cat√©gorie", "Impact√© √† Caps", "Impact√© √† Emilian"]
df = data[columns]
```

```python
import pandas as pd
import plotly.express as px
import dash
from dash import dcc, html
from dash.dependencies import Input, Output

# Initialiser l'application Dash
app = dash.Dash(__name__)

# Liste unique des ann√©es et des mois pour les filtres
ann√©es = df['Ann√©e'].unique()
mois = df['Mois'].unique()
personnes = ['Caps', 'Emilian']

# D√©finir la disposition de l'application avec des filtres interactifs
app.layout = html.Div(children=[
    html.H1(children='Tableau de Bord des D√©penses'),

    html.Div([
        dcc.Dropdown(
            id='ann√©e-dropdown',
            options=[{'label': str(ann√©e), 'value': ann√©e} for ann√©e in ann√©es],
            value=ann√©es[0],
            placeholder="S√©lectionnez une ann√©e"
        ),
        dcc.Dropdown(
            id='mois-dropdown',
            options=[{'label': m.capitalize(), 'value': m} for m in mois],
            value=mois[0],
            placeholder="S√©lectionnez un mois"
        ),
        dcc.Dropdown(
            id='personne-dropdown',
            options=[{'label': p, 'value': p} for p in personnes],
            value=personnes[0],
            placeholder="S√©lectionnez une personne"
        )
    ], style={'display': 'flex', 'gap': '10px'}),

    dcc.Graph(id='graph-depenses-mensuelles'),
    dcc.Graph(id='graph-depenses-categorie'),
    dcc.Graph(id='graph-depenses-personne')
])

# Callback pour mettre √† jour les graphiques en fonction des filtres
@app.callback(
    [Output('graph-depenses-mensuelles', 'figure'),
     Output('graph-depenses-categorie', 'figure'),
     Output('graph-depenses-personne', 'figure')],
    [Input('ann√©e-dropdown', 'value'),
     Input('mois-dropdown', 'value'),
     Input('personne-dropdown', 'value')]
)
def update_graphs(selected_year, selected_month, selected_person):
    filtered_df = df[(df['Ann√©e'] == selected_year) & (df['Mois'] == selected_month) & (df['Type de transaction'] == "D√©pense")]

    # Utiliser la colonne appropri√©e en fonction de la personne s√©lectionn√©e
    montant_colonne = f'Impact√© √† {selected_person}'

    
    # D√©penses par cat√©gorie
    depenses_par_categorie = filtered_df.groupby('Cat√©gorie')[montant_colonne].sum().reset_index()
    fig_categorie = px.pie(depenses_par_categorie, names='Cat√©gorie', values=montant_colonne, title='R√©partition des D√©penses par Cat√©gorie')

    # D√©penses par personne (fixe car filtr√© par personne)
    depenses_par_personne_par_mois = df.groupby('Mois')[montant_colonne].sum().reset_index()
    fig_personne = px.bar(depenses_par_personne_par_mois, x='Mois', y=montant_colonne, title='D√©penses par Personne')

    return fig_categorie, fig_personne

# Lancer l'application
if __name__ == '__main__':
    app.run_server(debug=True)
```
